```python
import json
import torch
import torch.nn.functional as F
import numpy as np
from torch_geometric.data import Data
from torch_geometric.nn import SAGEConv
from sklearn.preprocessing import MultiLabelBinarizer, LabelEncoder
import time
start = time.time()

# --- Load and preprocess data ---
with open('10kvalues.json', 'r') as f:
    data = json.load(f)

# Flatten all users (founders, investors, etc.) into one list
users = []
for key in data:
    users.extend(data[key])

users = [u for u in users if "name" in u]

# Assign each user a unique index
for idx, user in enumerate(users):
    user['idx'] = idx

# --- Feature Engineering ---
# Example: Use domains, role, and seeking/collaborators as features
domains = [tuple(u.get('domain', [])) for u in users]
roles = [u['role'] for u in users]
collab = [tuple(u.get('collaboratorsNeeded', [])) for u in users]
seeking = [tuple(u.get('seeking', [])) if isinstance(u.get('seeking', []), list) else (u.get('seeking', ''),) for u in users]

mlb_domains = MultiLabelBinarizer().fit(domains)
mlb_collab = MultiLabelBinarizer().fit(collab)
mlb_seeking = MultiLabelBinarizer().fit(seeking)
le_roles = LabelEncoder().fit(roles)

X_domains = mlb_domains.transform(domains)
X_roles = le_roles.transform(roles).reshape(-1, 1)
X_collab = mlb_collab.transform(collab)
X_seeking = mlb_seeking.transform(seeking)  # ✅ fixed here

# Concatenate all features
X = np.concatenate([X_domains, X_roles, X_collab, X_seeking], axis=1)
X = torch.tensor(X, dtype=torch.float)


# === Define complementary role mapping ===
def get_complementary_roles(user_role):
    complementary_mapping = {
        "founder": ["investor", "accelerator", "mentor", "legal", "cybersecurity", "mentors_research"],
        "investor": ["founder", "accelerator", "legal"],
        "accelerator": ["founder", "investor", "mentor", "legal"],
        "mentor": ["founder", "accelerator", "executive"],
        "legal": ["founder", "investor", "accelerator", "executive"],
        "government": ["founder", "accelerator"],
        "cybersecurity": ["founder", "executive"],
        "mentors_research": ["founder", "executive"],
        "executive": ["mentor", "legal", "cybersecurity", "freelancer"],
        "freelancer": ["founder", "executive"]
    }
    return complementary_mapping.get(user_role.lower(), [])

def roles_compatible(user1, user2):
    """Check if user1's role is compatible with user2's role (either direction)"""
    role1 = user1.get("role", "").lower()
    role2 = user2.get("role", "").lower()

    return (role2 in get_complementary_roles(role1)) or (role1 in get_complementary_roles(role2))


# --- Build Edges: Fully connect all users (no self-loops) ---
edge_index = []

for i in range(len(users)):
    for j in range(len(users)):
        if roles_compatible(users[i], users[j]):
            edge_index.append([i, j])
            edge_index.append([j, i])

edge_index = torch.tensor(edge_index, dtype=torch.long).t().contiguous()



# --- PyG Data Object ---
data = Data(x=X, edge_index=edge_index)

#  --- Define GNN Model ---
class MatchGNN(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels):
        super().__init__()
        self.conv1 = SAGEConv(in_channels, hidden_channels)
        self.conv2 = SAGEConv(hidden_channels, out_channels)

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = self.conv2(x, edge_index)
        return x
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

model = MatchGNN(in_channels=X.shape[1], hidden_channels=32, out_channels=16)


# --- "Training" (unsupervised, just get embeddings) ---
model.eval()
with torch.no_grad():
   embeddings = model(data.x, data.edge_index)




# --- Matchmaking: Find top-3 similar users for each user (cosine similarity) ---
from sklearn.metrics.pairwise import cosine_similarity

emb_np = embeddings.numpy()
similarities = cosine_similarity(emb_np)


# match_results = []
# for i, user in enumerate(users):
#     sim_scores = similarities[i]

#     # Find users with different roles only
#     other_roles = [idx for idx in np.argsort(sim_scores)[::-1] if users[idx]['role'] != user['role'] and idx != i]

#     top_idx = other_roles[:5]  # Top 3 cross-role matches

#     matches = [{
#         "name": users[idx]["name"],
#         "role": users[idx]["role"],
#         "similarity": float(sim_scores[idx])
#     } for idx in top_idx]

#     match_results.append({
#         "user": user["name"],
#         "role": user["role"],
#         "matches": matches
#     })
# seek_to_role = {
#     "Investments": "Investor",
#     "Partnerships": "Freelancer",
#     "Mentorship": "Mentor",
#     "Acceleration": "Accelerator",
#     "Tech Support": "Freelancer",
#     "Legal Advice": "Freelancer",
#     "Strategy": "Mentor",
#     "Hiring": "Freelancer"
# }


# match_results = []

# for i, user in enumerate(users):
#     sim_scores = similarities[i]
#     user_role = user["role"]

#     seeking = user.get("seeking", [])
#     if isinstance(seeking, str):
#         seeking = [seeking]
#     elif not seeking:
#         seeking = ["Investments", "Mentorship"]

#     seen = set()
#     matches = []

#     for keyword in seeking:
#         target_role = seek_to_role.get(keyword)
#         if not target_role:
#             continue  # skip unknown terms

#         count = 0
#         for idx in np.argsort(sim_scores)[::-1]:
#             if idx == i:
#                 continue
#             if users[idx]['role'] == target_role and users[idx]['role'] != user_role and idx not in seen:
#                 matches.append({
#                     "name": users[idx]["name"],
#                     "role": users[idx]["role"],
#                     "similarity": float(sim_scores[idx])
#                 })
#                 seen.add(idx)
#                 count += 1
#             if count == 3:
#                 break

#     match_results.append({
#         "user": user["name"],
#         "role": user["role"],
#         "seeking": seeking,
#         "matches": matches
#     })




from sklearn.metrics.pairwise import cosine_similarity
import json


# === Optional: helper to extract company or group ===
def get_company_identifier(user):
    return user.get("company", "") or user.get("org", "") or ""

# === Step: Compute cosine similarity matrix ===
emb_np = embeddings.numpy()
similarities = cosine_similarity(emb_np)

# === Step: Match top N for each user with complementary roles ===
NUM_MATCHES = 3
match_results = []

for user_idx, user in enumerate(users):
    sim_scores = similarities[user_idx]
    user_role = user.get("role", "").lower()
    allowed_roles = get_complementary_roles(user_role)

    if not allowed_roles:
        continue

    seen_roles = set()
    seen_companies = set()
    current_company = get_company_identifier(user)
    if current_company:
        seen_companies.add(current_company)

    matches = []

    for idx in np.argsort(sim_scores)[::-1]:
        if idx == user_idx:
            continue

        match = users[idx]
        match_role = match.get("role", "").lower()
        match_company = get_company_identifier(match)

        if match_role not in allowed_roles:
            continue
        if match_company and match_company in seen_companies:
            continue
        if match_role in seen_roles:
            continue

        matches.append({
            "name": match.get("name", ""),
            "role": match.get("role", ""),
            "company": match_company,
            "similarity": round(float(sim_scores[idx]), 4)
        })

        seen_roles.add(match_role)
        if match_company:
            seen_companies.add(match_company)

        if len(matches) == NUM_MATCHES:
            break

    match_results.append({
        "user": user.get("name", ""),
        "role": user.get("role", ""),
        "company": current_company,
        "matches": matches
    })

    if (user_idx + 1) % 50 == 0:
        print(f"Processed {user_idx + 1}/{len(users)} users")

# === Save to JSON
with open("gnn_complementary_matches.json", "w") as f:
    json.dump(match_results, f, indent=2)

print("✅ Matchmaking saved to gnn_complementary_matches.json")

# your code block (e.g., model.forward())
end = time.time()
print("Time taken:", end - start)
```
